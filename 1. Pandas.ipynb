## Pandas Co relations:


- corr() finds the relationship between only the numeric columns 
- non numeric columns are ignored.

```python
df11=pd.read_csv("data2.csv")
df11.corr()
```

### corr() output explaination:

- The numbers varies from -1 to +1.
- 1 means: there is 1 to 1 relationship. perfect co relation.
- Ex: pulse vs pulse. column vs same column has correlation 1
- Directly proportional: 1 or 0.9 is directly proportional.
- ex: if duration increases, calorie increases. 
- More duration I workout, more calorie is burnt. 
- If corr is 0.2 then there is no good corelation. 
- That is, if one goes up: other need not to go up or down. Neither proportional nor inversely proportional
- If the numbers are in negative: 
- Ex: -0.9: if one increases, other decreases. 
- Ex: speed and distance.if speed increases, distance decreases


#### Perfect Co relation:
- if the value is 1.000 ex: duration vs duration.
- make sense as each column has perfect relationship between itself

#### Good Co relation
- if the value is 0.99989. It is good co relation.
- ex: if I workout for more duration, more calories are burnt. 
- same with -0.99. Inversely proportional. 
- Both context means that there is relationship between columns

#### Bad Co relation
- 0.00089. Bad 
-  There is no increase or decrease in value with one another.
- we cant understand the relationship between 2 columns.
